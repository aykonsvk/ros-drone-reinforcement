```python
### One AXIS First
y_qlearn_w = pd.read_csv('../training_results/qlearn/y/Old-Wrong_params/logs.csv')
plt.figure(figsize=(14,7))
plt.plot(y_qlearn_w['episode'].rolling(10, min_periods=1).mean(),y_qlearn_w['steps'].rolling(10, min_periods=1).mean())
plt.plot(y_qlearn_w['episode'].rolling(10, min_periods=1).mean(),y_qlearn_w['reward'].rolling(10, min_periods=1).mean())
plt.title('One-axis movement')
plt.xlabel('Episodes')
plt.ylabel('Steps')
plt.legend(['Q-Learning Steps SMA(10)', 'Q-Learning Rewards SMA(10)'])
plt.savefig('one-axis-first.png')
plt.show()
```

## Model visualisation

```python
model = tf.keras.Sequential([

    Input((10)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(7)
])
model.compile(loss='mse', optimizer=Adam(0.001))

visualizer(model, format='png', view=True)
```
